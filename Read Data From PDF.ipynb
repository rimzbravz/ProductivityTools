```python
# Read every File:
    # Stack Dataframe in Every file
    
import os
filenames = os.listdir(r'C:\Users\bravrima\Downloads\PDFtoCSV')
print(filenames)
```

```python
# Initialize DataFrame:

#df_conso = pd.DataFrame()
df_for_conso = []

for i in filenames:
    print('Processing: '+i)
    df = tabula.read_pdf('C:\\Users\\bravrima\\Downloads\\PDFtoCSV\\'+str(i), pages='all')
    df_stack = pd.concat(df) #df stack is the dataframe in one file:
    print(df_stack.shape)
    
    df_for_conso.append(df_stack)

df_conso = pd.concat(df_for_conso) #Append to the initialized dataset tp gather it all
    

df_conso.shape

    
```
